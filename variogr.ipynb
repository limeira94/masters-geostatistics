```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Supondo que você tenha os dados do variograma experimental em arrays
distances = np.array([100, 200, 300, 400])
variogram_exp = np.array([1.51, 3.47, 4.31, 6.35])  # Substitua pelos valores calculados

# Definição da função do modelo exponencial
def exponential_variogram(h, nugget, sill, range):
    return nugget + sill * (1 - np.exp(-h / range))

# Estimativas iniciais para os parâmetros
initial_params = [0.1, 1.0, 300]

# Ajuste da curva
params, covariance = curve_fit(exponential_variogram, distances, variogram_exp, p0=initial_params)

# Parâmetros ajustados
nugget, sill, range = params

# Impressão dos parâmetros ajustados
print(f"Nugget: {nugget}, Sill: {sill}, Range: {range}")

# Plotagem dos dados experimentais e do modelo ajustado
plt.figure(figsize=(10, 6))
plt.scatter(distances, variogram_exp, color='blue', label='Variograma Experimental')
plt.plot(distances, exponential_variogram(distances, *params), color='red', label='Modelo Exponencial')
plt.xlabel('Distância')
plt.ylabel('Semivariância')
plt.legend()
plt.title('Ajuste do Variograma Teórico')
plt.show()
```
